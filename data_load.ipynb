```python
line_count = 0
count_pd = 0
sample_label = {}
with open('sample_list.txt', 'r') as f:
    for line in f:
        line_data = line.rstrip()
        line_count += 1
        if line_count == 2:
            sample_id = line_data[24:-17]
            print(line_data[24:-17])
        elif line_count == 14:
            label = line_data[19:-4]
            if label =='PD':
                count_pd += 1
            print(line_data[19:-4])
        elif line_count == 30:
            line_count = 0
        sample_label[sample_id] = label
```

```python
dir = 'PRJEB17784/'
list = ['ERR1912946']
string = ''
for run_id in list:
    for file_name in os.listdir(os.path.join(dir,run_id)):
        if os.path.splitext(file_name)[1] != '.gz':
            continue
        file_path = os.path.join(dir, run_id,file_name)
        string += file_path
        string += ','
```
